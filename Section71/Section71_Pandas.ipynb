## `597강`
- 값 분류 및 열 추가하기: 잠재력이 가장 큰 전공 대 위험도가 가장 낮은 전공




```python
df['Mid-Career 90th Percentile Salary'] - df['Mid-Career 10th Percentile Salary']
```

```python
df['Mid-Career 90th Percentile Salary'].subtract(df['Mid-Career 10th Percentile Salary'])
```

```python
spread = df['Mid-Career 90th Percentile Salary']- df['Mid-Career 10th Percentile Salary']
df.insert(1, 'gap', spread)
df.head(3)
```

```python
# 가장 낮은 분포로 정렬

low_sort_df = df.sort_values('gap')
low_sort_df[['Undergraduate Major', 'gap']].head()
```

## `598강`
- 해설 :잠재력이 가장 큰 학위

- 잠재력이 가장 높은 학위

```python
df.sort_values(by='Mid-Career 90th Percentile Salary', ascending=False).head(5)
```

```python
df.sort_values(by='Mid-Career 90th Percentile Salary', ascending=False).head(5).iloc[:5][['Undergraduate Major', 'Mid-Career 90th Percentile Salary']]
```

- 급여에서 가장 큰 분포를 보이는 학위

```python
high_grade = df.sort_values(by='gap', ascending= False).reset_index(drop=True)
high_grade.head(5)
```

```python
high_grade.iloc[:1]['Undergraduate Major']
```

- 중간봉급에서 가장 큰 분포를 보이는 학위

```python
df.sort_values(by='Mid-Career Median Salary', ascending=False).iloc[:5][['Undergraduate Major', 'Mid-Career Median Salary']]
```
