## `644강`
ndarray로 이미지 다루기

```python
import numpy as np
import matplotlib.pyplot as plt


from scipy import misc
from PIL import Image
```

```python
img = misc.face()
plt.imshow(img);
```

```python
type(img)
```

```python
print(img.shape)
print(img.ndim)
```

```python
# 이미지를 그레이스케일로 변환

# 1. 스칼라로 나눔
sRGB_array = img/255
print(sRGB_array)

# 2. Numpy의 브로드캐스팅을 사용해 ndarray의 모든값을 255로 나움
grey_vals = np.array([0.2126, 0.7152, 0.0722])

# 3. 공식에 의해 주어진 값 구하기
img_gray = sRGB_array @ grey_vals
print(img_gray)

# 4. @ 연산자 혹은 .matmul() 함수를 통해 곱함
# img_gray = sRGB_array @ grey_vals
img_gray = np.matmul(sRGB_array, grey_vals)
print(img_gray)

# 5. 이미지 표시를 위해 Matplotlib 사용
plt.imshow(img_gray, cmap='gray')
```

```python
plt.imshow(img_gray)
```

```python
plt.imshow(np.flip(img_gray), cmap='gray')
```

```python
plt.imshow(np.rot90(img_gray), cmap='gray')
```
