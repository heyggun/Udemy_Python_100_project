## `644`
ndarray  

```python
import numpy as np
import matplotlib.pyplot as plt


from scipy import misc
from PIL import Image
```

```python
img = misc.face()
plt.imshow(img);
```

```python
type(img)
```

```python
print(img.shape)
print(img.ndim)
```

```python
#   

# 1.  
sRGB_array = img/255
print(sRGB_array)

# 2. Numpy   ndarray  255 
grey_vals = np.array([0.2126, 0.7152, 0.0722])

# 3.     
img_gray = sRGB_array @ grey_vals
print(img_gray)

# 4. @   .matmul()   
# img_gray = sRGB_array @ grey_vals
img_gray = np.matmul(sRGB_array, grey_vals)
print(img_gray)

# 5.    Matplotlib 
plt.imshow(img_gray, cmap='gray')
```

```python
plt.imshow(img_gray)
```

```python
plt.imshow(np.flip(img_gray), cmap='gray')
```

```python
plt.imshow(np.rot90(img_gray), cmap='gray')
```
